```python
import textwrap
import shutil

margem = 2
largura_terminal = 50
largura_util = largura_terminal - 2 * margem

texto = "Texto centralizado com margens."
linhas = textwrap.wrap(texto, width=largura_util)
for linha in linhas:
    print(" " * margem + linha.center(largura_util) + " " * margem)


def diagnostico_lesoes_proliferativas():
    """
    Função para diagnóstico de lesões proliferativas na mucosa bucal
    """

    print("🦷 Diagnóstico de Lesões Proliferativasas na Mucosa Bucal 🦷\n")

    # Passo 1: Determinar a natureza da lesão
    natureza = input("🤔 A lesão é reacional ou tumoral? (reacional/tumoral): ").lower()

    if natureza == "reacional":
        print("\n🔬 Lesão benigna (reacional)\n")

        # Passo 2: Determinar a localização preferencial da lesão reacional
        localizacao = input("📍 Qual a localização preferencial da lesão? (labio inferior, assoalho bucal, palato, rebordo, mucosa jugal, gengiva, outras): ").lower()

        if localizacao == "labio inferior":
            print("\n➡️  Diagnóstico provável: MUCOCELE")
            print("     - 💧 **Definição e Etiologia:** Ruptura de ducto salivar levando ao acúmulo de mucina.")
            print("     - 📍 **Localização:** Lábio inferior (principal).")
            print("     - 🔵 **Aspecto Clínico:**  Histórico de recorrência. Tumefação com coloração translúcida/azulada ou semelhante à mucosa, com consistência fibrosa ou flutuante. Pode romper espontaneamente.")
            print("     - 🩹 **Conduta:** Remoção da lesão e das glândulas salivares menores associadas.\n")

        elif localizacao == "assoalho bucal":
            print("\n➡️  Diagnóstico provável: RÂNULA")
            print("     - 💧 **Definição e Etiologia:** Extravasamento de muco da glândula sublingual.")
            print("     - 📍 **Localização:** Assoalho bucal.")
            print("     - 🔵 **Aspecto Clínico:** Tumefação com coloração translúcida/azulada ou semelhante à mucosa. Consistência fibrosa ou flutuante. Pode elevar a língua.")
            print("     - 🩹 **Conduta:** Marsupialização ou remoção da glândula sublingual.\n")

        elif localizacao in ["palato", "rebordo", "mucosa jugal"]:
            print("\n➡️  Diagnóstico provável: HIPERPLASIA INFLAMATÓRIA")
            print("     - 🤕 **Definição e Etiologia:** Associado à trauma ou fator irritativo crônico (próteses desadaptadas, dentes fraturados, hábitos nocivos).")
            print("     - 📍 **Localização:** Palato, rebordo e mucosa jugal.")
            print("     - ⚪ **Aspecto Clínico:** Crescimento lento, coloração semelhante à mucosa, superfície lisa ou irregular, tamanho variável.")
            print("     - 🔬 **Conduta:** Biópsia excisional seguida de exame histopatológico. Eliminação do fator irritante.\n")

        elif localizacao == "gengiva":
            # Passo 3:  Diferenciar entre FOP e LCG
            deslocamento_dentes = input("🦷 A lesão causa deslocamento dos dentes? (sim/nao): ").lower()
            if deslocamento_dentes == "sim":
                print("\n➡️  Diagnóstico provável: FIBROMA OSSIFICANTE PERIFÉRICO (FOP)")
                print("     - 💪 **Definição e Etiologia:** Tecido mineralizado.")
                print("     - 📍 **Localização:** Gengiva.")
                print("     - ⚪ **Aspecto Clínico:** Coloração semelhante à mucosa ou esbranquiçada, superfície lisa ou lobulada, séssil ou pediculada, consistência fibrosa ou endurecida. Pode causar deslocamento dos dentes.")
                print("     - 🔬 **Conduta:** Biópsia excisional seguida de exame histopatológico.\n")
            else:
                print("\n➡️  Diagnóstico provável: LESÃO PERIFÉRICA DE CÉLULAS GIGANTES (LCG)")
                print("     - 🩸 **Definição e Etiologia:** Reação associada a trauma ou inflamação.")
                print("     - 📍 **Localização:** Gengiva.")
                print("     - 🔴 **Aspecto Clínico:** Coloração avermelhada ou vermelho-azulada, sangrante ao toque, superfície lisa ou irregular, base séssil ou pediculada.")
                print("     - 🔬 **Conduta:** Biópsia excisional seguida de exame histopatológico.\n")

        elif localizacao == "outras":
            print("\n➡️  Diagnóstico provável: GRANULOMA PIOGÊNICO")
            print("     - 🩸 **Definição e Etiologia:** Lesão vascularizada, reação a trauma ou irritação local.")
            print("     - 📍 **Localização:** Gengiva e outras localizações.")
            print("     - 🔴 **Aspecto Clínico:** Coloração avermelhada, superfície friável, base séssil ou pediculada.")
            print("     - 🔬 **Conduta:** Biópsia excisional seguida de exame histopatológico.\n")

        else:
            print("\n🙁 Não foi possível determinar um diagnóstico entre as lesões reacionais com as informações fornecidas.")
            print("ℹ️  Considere a possibilidade de uma lesão maligna.\n")

    elif natureza == "tumoral":
        print("\n🔬 Lesão benigna (tumoral)\n")

        # Passo 4: Determinar a localização preferencial da lesão tumoral
        localizacao = input("📍 Qual a localização preferencial da lesão? (mucosa jugal, vestibulo, lingua, outras): ").lower()

        if localizacao in ["mucosa jugal", "vestibulo"]:
            # Passo 5: Diferenciar entre Lipoma e Fibroma
            cor = input("🎨 Qual a cor da lesão? (amarelada/outra cor): ").lower()
            if cor == "amarelada":
                print("\n➡️  Diagnóstico provável: LIPOMA")
                print("     - 🟡 **Definição e Etiologia:** Neoplasia benigna de adipócitos.")
                print("     - 📍 **Localização:** Mucosa jugal e vestíbulo.")
                print("     - 🟡 **Aspecto Clínico:** Lesão nodular, bem delimitada, crescimento lento, séssil ou pediculada, superfície lisa, coloração amarelada ou semelhante à mucosa.")
                print("     - 🔬 **Conduta:** Biópsia excisional seguida de exame histopatológico.\n")
            else:
                print("\n➡️  Diagnóstico provável: FIBROMA")
                print("     - ⚪ **Definição e Etiologia:** Neoplasia benigna de fibroblastos.")
                print("     - 📍 **Localização:** Mucosa jugal.")
                print("     - ⚪ **Aspecto Clínico:** Lesão nodular, bem delimitada, crescimento lento, séssil ou pediculada, superfície lisa, coloração semelhante à mucosa ou esbranquiçada.")
                print("     - 🔬 **Conduta:** Biópsia excisional seguida de exame histopatológico.\n")

        elif localizacao == "lingua":
             # Passo 6: Diferenciar entre Hemangioma e Linfangioma
            cor = input("🎨 Qual a cor da lesão? (arroxeada/semelhante a mucosa ou avermelhada): ").lower()
            if cor == "arroxeada":
                diascopia = input("🩸 A lesão mostra-se isquemiada à diascopia? (sim/nao): ").lower()
                if diascopia == "sim":
                    print("\n➡️ Diagnóstico provável: HEMANGIOMA")
                    print(" - 🩸 **Definição e Etiologia:** Proliferação de vasos sanguíneos.")
                    print(" - 📍 **Localização:** Pode ocorrer em várias localizações na mucosa bucal.")
                    print(" - 🔴 **Aspecto Clínico:** Limites mal definidos, superfície irregular, coloração arroxeada, mostra-se isquemiada à diascopia.")
                    print(" - 💉 **Conduta:** Escleroterapia, se necessário, ou acompanhamento (evitar biópsia por risco de hemorragia).\n")

            else:
                print("\n➡️  Diagnóstico provável: LINFANGIOMA")
                print("     - 🌊 **Definição e Etiologia:** Proliferação de vasos linfáticos.")
                print("     - 📍 **Localização:** Língua.")
                print("     - ⚪ **Aspecto Clínico:** Limites mal definidos, superfície pedregosa, coloração semelhante à mucosa ou avermelhada.")
                print("     - 🔬 **Conduta:** Biópsia incisional se necessário. Acompanhamento.\n")

        elif localizacao == "outras":
           # Passo 7: Diferenciar entre Papiloma e Tumor Benigno de Glândula Salivar
            superficie = input("🔬 A lesão apresenta superfície exofítica? (sim/nao): ").lower()
            if superficie == "sim":
                print("\n➡️ Diagnóstico provável: PAPILOMA")
                print(" - 🦠 **Definição e Etiologia:** Associado ao HPV.")
                print(" - 📍 **Localização:** Pode ocorrer em várias localizações na mucosa bucal.")
                print(" - ⚪ **Aspecto Clínico:** Projeções digitiformes, superfície exofítica.")
                print(" - 🔬 **Conduta:** Remoção total + análise histopatológica.\n")
            else:
                print("\n➡️ Diagnóstico provável: TUMOR BENIGNO DE GLÂNDULA SALIVAR")
                print(" - 🧪 **Definição e Etiologia:**  Neoplasias benignas que se originam nas glândulas salivares.")
                print(" - 📍 **Localização:** Pode ocorrer em várias localizações, incluindo glândulas salivares maiores e menores.")
                print(" - ⚪ **Aspecto Clínico:** Crescimento lento e expansivo, sem alteração de mucosa.")
                print(" - 🔬 **Conduta:** Remoção e análise histopatológica.\n")

        else:
            print("\n🙁 Não foi possível determinar um diagnóstico entre as lesões tumorais com as informações fornecidas.")
            print("ℹ️  Considere a possibilidade de uma lesão maligna.\n")

    else:
        print("\n🙁 Não foi possível determinar um diagnóstico para lesão proliferativa com as informações fornecidas.")
        print("ℹ️  Considere a possibilidade de uma lesão maligna.\n")

    # Informações sobre lesões malignas
    print("\n🔴 Possibilidade de Lesão Maligna (consultar um especialista para diagnóstico):")
    print("➡️ CARCINOMA ESPINOCELULAR (CEC)")
    print("     - 🚩 **Definição e Etiologia:** Tumor maligno de células escamosas.")
    print("     - 📍 **Localização:** Língua e assoalho bucal.")
    print("     - 🔴 **Aspecto Clínico:** Crescimento rápido, massa exofítica de consistência firme, superfície fungiforme, papilar ou verrucosa, coloração semelhante à mucosa, avermelhada ou esbranquiçada.")
    print("     - 🔬 **Conduta:** Biópsia incisional seguida de exame histopatológico.\n")

    print("➡️ TUMOR MALIGNO DE GLÂNDULA SALIVAR")
    print("     - 🚩 **Definição e Etiologia:** Tumor maligno de glândulas salivares.")
    print("     - 📍 **Localização:** Palato e mucosa labial.")
    print("     - 🔴 **Aspecto Clínico:** Crescimento pode ser rápido ou lento, coloração semelhante à da mucosa normal adjacente ou azulada, mucosa íntegra ou ulcerada, assintomática ou com dor espontânea.")
    print("     - 🔬 **Conduta:** Biópsia incisional seguida de exame histopatológico.\n")

    print("\n❗ Importante: Encaminhamento ao cirurgião de cabeça e pescoço imediatamente!\n")

diagnostico_lesoes_proliferativas()
```
